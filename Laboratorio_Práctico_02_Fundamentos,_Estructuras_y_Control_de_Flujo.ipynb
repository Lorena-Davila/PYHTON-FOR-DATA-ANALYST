# Caso de Negocio: Limpieza y Clasificación de Transacciones

Usted forma parte del equipo de Ingeniería de Datos de una pasarela de pagos. El sistema
heredado (legacy) envía los registros de las transacciones diarias en un formato de texto crudo
y desordenado. Su tarea es crear un script que reciba este registro, lo limpie, evalúe el nivel de
riesgo de la transacción y consolide la información en un formato estructurado para el reporte
de cierre.

### Parte 1: Ingesta y Limpieza de Cadenas (Strings)


```python
# Utilice el método adecuado para eliminar los espacios en blanco y los símbolos de dólar
# $$ que se encuentran en los extremos de la cadena.

log_crudo = "$$[ESTADO:ok]||trx_id:9934||monto:S/.1250.50||alerta:falso$$ "

step_1 = log_crudo.strip("$ ")
step_1
```

```python
# Convierta toda la cadena resultante a letras mayúsculas.
step_2 = step_1.upper()
print(step_2)
```

```python
# El delimitador que separa cada campo es la doble barra vertical ||.
# Utilice el método correcto para dividir la cadena en partes usando este delimitador.

step_3 = step_2.split('||')
step_3
```

```python
# Guarde el resultado final en una variable llamada lista_campos e imprímala en pantalla
# usando print().

lista_campos = step_3
print(lista_campos)
```

### Parte 2: Colecciones y Extracción Analítica

```python
# Acceda al tercer elemento de su lista_campos (recuerde cómo funcionan los índices en
# Python) donde se encuentra la información del monto.

lista_campos[2]
```

```python
# Utilizando técnicas de Slicing (rebanado) o métodos de strings, extraiga únicamente la
# porción numérica (es decir, excluya el texto "MONTO:S/.").

monto_limpio = lista_campos[2].replace("MONTO:S/.", "")
monto_limpio
```

```python
# Convierta ese texto extraído a un tipo de dato decimal (float) y guárdelo en una variable
# llamada monto_transaccion.

monto_transaccion = float(monto_limpio)
monto_transaccion
```

### Parte 3: Lógica Condicional y Tuplas de Control

El departamento de Prevención de Fraude tiene umbrales estrictos para auditar transacciones.
Estos umbrales no deben ser modificados por error durante la ejecución del programa.

```python
umbrales_auditoria = (500.0, 2000.0)

print(f"El límite de revisión manual es {umbrales_auditoria[0]}")
print(f"El límite de bloqueo por presunto fraude es {umbrales_auditoria[1]}")
```

```python
# Implemente una estructura condicional (if - elif - else) que evalúe la variable
# monto_transaccion:

print(f"El monto de transaccion es {monto_transaccion}")

# Si el monto es estrictamente mayor al límite de bloqueo (posición 1 de la tupla),
# imprima: "ALERTA ROJA: Transacción bloqueada. Supera los $2000."

if monto_transaccion > umbrales_auditoria[1]:
  print("ALERTA ROJA: Transacción bloqueada. Supera los $2000.")

# Si el monto es mayor al límite de revisión
elif umbrales_auditoria[0] < monto_transaccion <= umbrales_auditoria[1]:
  print("ALERTA AMARILLA: Transacción rechazada. Supera los $500.")

# Si no se cumple ninguna de las anteriores
else:
  print("VERDE: Transacción procesada automáticamente.")
```

# Parte 4: Diccionarios y Automatización con Bucles

Para enviar la información al sistema de reportes, debemos consolidarla en un diccionario e
iterar sobre los resultados.

```python
# 1. Cree un Diccionario llamado reporte_diario con las siguientes llaves y valores:

montos = [350.0, 1250.50, 90.0, 2100.0]
nombre_apellido = input("Ingrese su nombre y apellido: ")

reporte_diario = {"fecha": "2026-06-07",
                  "lotes_procesados": montos,
                  "operador": nombre_apellido}
```

```python
# 2. Utilice un bucle for combinado con la función enumerate() para recorrer la lista de
# lotes_procesados dentro del diccionario.

for indice, montos in enumerate(reporte_diario["lotes_procesados"], start = 1):
    print(indice, montos)
```

```python
# 3. Dentro del bucle, utilice una f-string para imprimir: "Lote N° [Índice] procesado por un
# valor de $[Monto]". Haga que el índice comience en 1.

for indice, montos in enumerate(reporte_diario["lotes_procesados"], start = 1):
    print(f"Lote N° {indice} procesado por un valor de ${montos}.")
```

```python
# 4. Simule un proceso de conexión de red inestable utilizando un bucle while. Configure un
# contador de intentos inicializado en 1. El bucle debe imprimir "Intentando conectar al
# servidor de destino... Intento [X]" y detenerse automáticamente al llegar al intento número

tope = 3
numero_intentos = 1

while numero_intentos <= tope:
    print(f"Intentando conectar al servidor de destino... Intento {numero_intentos}")
    numero_intentos = numero_intentos + 1 # contador
```
