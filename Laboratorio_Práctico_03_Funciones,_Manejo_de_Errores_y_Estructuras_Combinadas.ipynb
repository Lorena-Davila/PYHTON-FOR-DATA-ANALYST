# Caso de Negocio: Desarrollo de un Módulo de Extracción Segura

Usted es el encargado de construir una "fábrica" de código (Funciones) para procesar los datos
que llegan desde la pasarela de pagos web (API). La información llega en forma de una Lista
de Diccionarios, pero algunos registros están corruptos. Su objetivo es extraer los montos,
convertirlos a moneda local esquivando los errores del sistema, y generar una lista de auditoría
en tiempo récord utilizando List Comprehensions.

## Parte 1: Estructuras Combinadas y Funciones Básicas
El servidor nos envía un bloque de datos JSON que en Python se lee como una Lista de
Diccionarios. Copie la siguiente variable en su código:

```python
respuesta_api = [
{"trx_id": "T001", "monto_usd": 150.00, "estado":
"Completado"},
{"trx_id": "T002", "monto_usd": "Error_Red", "estado":

"Fallido"},
{"trx_id": "T003", "monto_usd": 320.50, "estado":
"Completado"},
{"trx_id": "T004", "monto_usd": 45.00, "estado": "Completado"}
]
```

Misiones:
1. Cree una función llamada extraer_montos que reciba como parámetro una lista de datos
(como respuesta_api).
2. Dentro de la función, inicialice una lista vacía llamada lista_montos.
3. Utilice un bucle for para recorrer la lista ingresada. En cada ciclo, extraiga el valor de la
llave "monto_usd" y agréguelo a lista_montos utilizando el método .append().
4. La función debe retornar la lista final. Pruebe su función pasándole respuesta_api e
imprima el resultado.

```python
def extraer_montos(lista_datos):
  #inicialice una lista vacía
  lista_montos = []

  for dato in lista_datos:
    montos_limpios = dato['monto_usd']
    lista_montos.append(montos_limpios)

  return lista_montos

extraer_montos(respuesta_api)
```

## Parte 2: Manejo de Errores (Red de Seguridad)
Necesitamos convertir los montos extraídos a moneda local (Soles). Sin embargo, notará que
hay un texto intruso ("Error_Red") que hará que su programa colapse si intenta multiplicarlo
matemáticamente.

Misiones:
1. Cree una nueva función llamada convertir_a_soles que reciba un parámetro llamado
monto.
2. Implemente un bloque try - except dentro de la función.
3. En el bloque try, calcule la conversión multiplicando el monto por 3.8, redondee a 2
decimales y retorne ese resultado.
4. En el bloque except, capture la excepción general (Exception as e). Imprima un mensaje
de alerta (ej: "⚠️ Alerta: Dato corrupto detectado. Detalle: [Error]") y retorne el valor
numérico 0 para que el programa pueda continuar operando sin fallar.

```python
def convertir_a_soles(monto):
  # Usamos try/except para evitar que el programa se detenga
  # cuando encuentre un dato corrupto como "Error_Red".
  try:
      monto_convertido = round(monto * 3.8, 2)
      return monto_convertido

  except Exception as e:
      print(f"⚠️ Alerta: Dato corrupto detectado. Detalle: {e}")
      return 0
```

```python
# Procesamiento completo
lista_montos = extraer_montos(respuesta_api)

montos_convertidos = []

for monto in lista_montos:
    monto_en_soles = convertir_a_soles(monto)
    montos_convertidos.append(monto_en_soles)

print(montos_convertidos)
```

## Parte 3: List Comprehension (Procesamiento Ágil)
El área de Finanzas le ha enviado una lista ya limpia y validada en Soles, y le pide extraer
únicamente los impuestos de las transacciones consideradas "Grandes" en una sola línea de
código.

Misiones:
1. Utilice la técnica de List Comprehension para crear una nueva lista llamada
impuestos_grandes.
2. La lógica interna debe ser: Calcular el 18% (monto * 0.18) por cada monto en
montos_limpios, SOLO SI el monto es estrictamente mayor a 500.0.
3. Imprima la lista impuestos_grandes resultante y confirme que solo contiene 3 valores.

```python
montos_limpios = [570.0, 0, 1217.9, 171.0, 850.0, 90.0]
impuestos_grandes = [monto for monto in montos_limpios if monto > 500]
print(impuestos_grandes)
```
